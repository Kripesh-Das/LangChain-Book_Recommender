```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd 
books = pd.read_csv("cleaned_books.csv")
books
```

```python
books["tagged_description"]
```

```python
books["tagged_description"].to_csv("tagged_description.txt",sep="\n",index=False,header=False)
```

```python
raw_docs = TextLoader("tagged_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
docs = text_splitter.split_documents(raw_docs)
```

```python
docs[0]
```

```python
db_books = Chroma.from_documents(docs, embedding=OpenAIEmbeddings())
```

```python
query = "COMICS"
ans = db_books.similarity_search(query,k=5)
ans
```

```python
books[books["isbn13"]==int(ans[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book about superheros")
```
